```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re
import urllib.request
import urllib
from bs4 import BeautifulSoup, Comment
```

```python
url = 'https://www.wtatennis.com/tournaments'
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html,'lxml')
```

```python
active_tournaments = soup.find('div',{'class':
                                      'views-row views-row-1 views-row-odd views-row-first views-row-last'})
tournaments = active_tournaments.findAll('div',{'typeof': "sioc:Item foaf:Document"})
```

```python
for i in range(len(tournaments)):
    footer = tournaments[i].find('div',{'class': 'group-footer field-group-div'})
    footer_draw = footer.find('div',{'class':
                               'field field--name-draw-tournament-button field--type-ds field--label-hidden'})
    for a in footer_draw.findAll('a', href=True):
        draw_page_url = a['href']
    html_draw_page = urllib.request.urlopen(draw_page_url)
    soup_draw_page = BeautifulSoup(html_draw_page,'lxml')
    R64 = soup_draw_page.find('div',{'role': 'main'}).find('div',{'typeof': 'sioc:Item foaf:Document'}).find('div',
                                {'class': 'horizontal-tabs clearfix'})
    print(R64.text)
```

```python
comments = soup_draw_page.findAll(text=lambda text:isinstance(text, Comment))

# look for table with the id "team_stats"
rx = re.compile(r'<div.+?id="singles-draw-tab".+?>[\s\S]+?</table>')
for comment in comments:
    try:
        table = rx.search(comment.string).group(0)
        print(table)
        # break the loop if found
        break
    except:
        pass
```

```python
from bs4 import BeautifulSoup
import urllib.request
import re
comm = re.compile("<!--|-->")
def make_soup(url):
    page = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(comm.sub("", page.read()), 'lxml')
    return soupdata

def get_player_totals():
    soup = make_soup("http://www.pro-football-reference.com/years/2015/")

    tableStats = soup.find('table', {'id':'team_stats'})

    return tableStats

print(get_player_totals())
```
